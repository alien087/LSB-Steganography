```python
import math
from PIL import Image, ImageDraw
import tkinter as tk 
from tkinter import filedialog 
```

```python
root = tk.Tk() 
root.withdraw() 
file_path = filedialog.askopenfilename() 
image = Image.open(file_path)
ImageDraw.Draw(image)
width, height = image.size
```

```python
bit_image = list(image.getdata())
```

```python

test_str = "apa kabarnya semua temen temen sekalian, saya berharap teman teman semua sehat ya, oke ga nih? okeh dong"
byte_array = test_str.encode()

binary_int = int.from_bytes(byte_array, "big")
res = bin(binary_int)

res = res.replace('0b', '')
res
```

```python
bit_image[len(bit_image)-1] = (0,0,int((len(res)+1)/8))
print(bit_image)
```

```python
bit_images = []
for i in range (len(res)):
    if (res[i]!=format(bit_image[i][0], "08b")[7]):
        if(res[i]=="1"):
            bit_image[i] = bit_image[i][0]+1, bit_image[i][1], bit_image[i][2]
        elif(res[i]=="0"):
             bit_image[i] = bit_image[i][0]-1, bit_image[i][1], bit_image[i][2]
```

```python
image2 = Image.new(image.mode, (width, height))
image2.putdata(bit_image)
image2.save("image_cover.png")
```
