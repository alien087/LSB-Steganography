```python
image_cover = Image.open("image_cover.bmp")
image_cover=image_cover.convert("RGB")
ImageDraw.Draw(image_cover)
```

```python
bit_image2 = list(image_cover.getdata())
bit_image2[len(bit_image2)-1][2]
```

```python
kata = []
for i in range (((bit_image2[len(bit_image2)-1][2])*8)-1):
    kata.append(format(bit_image2[i][0], "08b")[7])
```

```python
a = ""
a = a.join(kata)
a
```

```python
binary_int = int(a, 2)
byte_number = binary_int.bit_length() + 7 // 8

binary_array = binary_int.to_bytes(byte_number, "big")
ascii_text = binary_array.decode()

print(ascii_text)
```
